```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1.scatterplot :scatter plot is used to visualize the relationship between two continuous variables.
2.barplot:Bar plots are effective for comparing categories or groups using rectangular bars
3.histplot:Histograms are used to visualize the distribution of a single variable
4.relplot:It is often used to visualize the relationship between two numerical variables. 
5.pairplot: The pairplot function in Seaborn creates a grid of subplots that displays pairwise relationships between multiple variables in a dataset.
```

```python
'''Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
'''

import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset('fmri')
sns.lineplot(x = fmri.timepoint, y = fmri.signal,data = fmri,hue = 'event',style='region')
plt.show()
```

```python
''' Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
'''

import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
plt.subplot(1,2,1)
sns.boxplot(x=titanic.pclass,y = titanic.age,data = titanic )
plt.title("Boxplot 1")

plt.subplot(1,2,2)
sns.boxplot(x=titanic.pclass,y = titanic.fare,data = titanic )
plt.title("Boxplot 2")

plt.show()
```

```python
''' Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.'''

import seaborn as sns
import matplotlib.pyplot as plt
diamond = sns.load_dataset('diamonds')
sns.histplot(data = diamond,x='price',hue = 'cut',kde  = True)
plt.show()

#sns.relplot(x= diamond.price,y = diamond.cut,data = diamond,hue = 'cut')
#plt.show()
```

```python
'''Q5.Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. '''
import seaborn as sns 
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.pairplot(data = iris, hue = 'species')
plt.show()
```

```python
''' 6: Use the "flights" dataset from seaborn to plot a heatmap. '''
import seaborn as sns 
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot('month','year','passengers')
sns.heatmap(flights_pivot)
plt.xlabel('Year')
plt.ylabel('Month')
plt.title(' Heatmap')
plt.show()
```
