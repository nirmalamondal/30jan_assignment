```python
'''
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
'''


'''A scatter plot is a type of data visualization that displays the relationship between two variables. It represents individual data points as markers on a two-dimensional
graph, with each point's position determined by the values of the two variables being plotted.'''


import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This x axis")
plt.ylabel("This y axis")
plt.title("This is scatterplot")
```

```python
'''
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
'''





'''The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure window into multiple smaller plots
or axes, where you can plot different data or visualize different aspects of the same data.
'''

x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1) 
plt.plot(x, y1)
plt.title('Line 1')


plt.subplot(2, 2, 2) 
plt.plot(x, y2)
plt.title('Line 2')


plt.subplot(2, 2, 3) 
plt.plot(x, y3)
plt.title('Line 3')


plt.subplot(2, 2, 4) 
plt.plot(x, y4)
plt.title('Line 4')
```

```python
'''
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
'''


'''  A bar plot, also known as a bar chart, is a type of data visualization that uses rectangular bars to represent categorical data. It displays the values of different 
categories as bars with lengths proportional to the values they represent. ar plots are commonly used for Comparison, Distribution,Ranking,Visualizing'''


import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

```python
'''
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
'''


"""A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution of a numerical dataset through quartiles. It provides a
summary of the dataset's central tendency, spread, and skewness.
box plots are used in Identifying Skewness and Outliers,Detecting Dispersion and Variability,Visualizing Distribution

"""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel("Values")
plt.title("box plot")
plt.show()
```
