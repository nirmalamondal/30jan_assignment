```python
#Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

'''ZeroDivisionError , FloatingPointError and OverflowError are defined in the ArithmeticError class.'''
import logging
logging.basicConfig(filename = "error.log",level = logging.ERROR)

#ZeroDivisionError

try:
    a = 10/0
except ZeroDivisionError as e:
    logging.error("This is handling {}".format(e))
    
''' Here zerodivision error is handling'''    

#OverflowError

j =3
try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    logging.error("This is handling {}".format(e))
    
''' Here result is very large so stack over flow is happen'''    
```

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
'''LookupError class is used when IndexError or KeyError exceptions are occcured.'''
import logging
logging.basicConfig(filename = "error.log",level = logging.ERROR)

#IndexError
try :
    l = [4,6,8,3,]
    print(l[8])
except IndexError as e:
    logging.error("This is handling {}".format(e))
    
    
#KeyError
try :
    d = {'a':5,1:[7,8]}
    print(d['key3'])
except KeyError as e:
    logging.error("This is handling {}".format(e))
    
```

```python
#Q5. Explain ImportError. What is ModuleNotFoundError?


'''ImportError handle the exception when we try to import a class that is not present in python library.'''

import logging
logging.basicConfig(filename = "error.log",level = logging.ERROR)
try :
    import nirmala
except ImportError as e:
    logging.error("This is handling {}".format(e))
    
'''ModuleNotFoundError occur when we try to access or use a module that cannot be found.'''

try :
    import numpys as numpy
except ModuleNotFoundError as e:
    logging.error("This is handling {}".format(e))
    
```
