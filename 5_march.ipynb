```python
#Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
'''
In Bokeh, glyphs are visual markers used to represent data points on a plot. They can be used to create various types of plots such as scatter plots, line plots, bar charts,
and more. Glyphs allow you to visually represent data in different ways, such as using circles, squares, triangles, or other shapes.
'''

from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title='Scatter Plot', x_axis_label='X-axis', y_axis_label='Y-axis')
#circle glyphs
p.circle(x, y, size=10, color='blue')
show(p)
```

```python
#Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
'''
To modify the axis labels, you can set the x_axis_label and y_axis_label properties of the figure object and modify their properties, such as axis_label_text_font_size,
major_label_text_color, major_tick_line_color, etc.
To set the title of the plot, you can use the title property of the figure object and can modify properties such as title_text_font_size, title_text_color, etc.
To add a legend to the plot, you can use the legend property of the figure object.Additional customization options include modifying properties such as legend_location,
 legend_label_text_font_size, legend_background_color, etc.
'''

from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title='Customized Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# Customize axes
p.xaxis.axis_label_text_font_style = 'italic'
p.yaxis.axis_label_text_font_style = 'italic'
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_font_size = '14pt'

# Customize title
p.title.text_font_size = '16pt'
p.title.text_color = 'blue'

p.square(x, y, size=10, color='blue', legend_label='Data Points')

# Customize legend
p.legend.location = 'top_left'
p.legend.label_text_font_size = '12pt'

show(p)
```
