```python
'''Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you might choose one over the other. '''


Ordinal Encoding is used when categories are have any order or ranking and otherside Label encoding assigns a label to each category without considering any 
ordinal relationship between the categories.

example:
'size' features have 'small','medium','large' categories which they be encoded as 0, 1, and 2, respectively in ordinal encoding.
and 'color' features have 'red','blue','green' categories which they be encoded as 3, 0, and 2, respectively without any order.
```

```python
'''Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine learning project.'''

Target Guided Ordinal Encoding is a technique used to transform categorical variables based on the relationship between the categories and the target variable
It assigns numerical labels to the categories, considering the target variable s average or aggregated value for each category. 
Target Guided Ordinal Encoding can be useful in a machine learning project when there is a significant relationship between the categorical variable and the 
target variable. It allows the model to capture the trend or pattern in the target variable corresponding to each category. This encoding technique can be 
particularly beneficial when there is a monotonic relationship between the categorical feature and the target variable.


Example: Consider a machine learning project where you are predicting the risk level of loan applicants (target variable) based on their credit history 
(categorical feature). The credit history can be divided into categories such as 'Good,' 'Fair,' and 'Poor.' By applying Target Guided Ordinal Encoding, you 
can assign numerical labels to these categories based on the average risk level associated with each credit history category. This encoding allows the model 
to consider the monotonic relationship between the credit history and the risk level, potentially improving the predictive performance.
```

```python
'''Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?'''

Covariance is a statistical measure that quantifies the relationship and degree of dependence between two random variables. It measures how changes in one 
variable are associated with changes in another variable.

1.Covariance helps to understand the relationship between two variables. A positive covariance suggests a direct relationship, meaning that as one variable 
increases, the other tends to increase as well. A negative covariance indicates an inverse relationship, meaning that as one variable increases, the other 
tends to decrease.
2.Covariance is useful for feature selection in statistical analysis and machine learning.
3.It plays a crucial role in finance and portfolio management. It helps measure the degree of dependency between different financial assets. Positive 
covariance between assets suggests they tend to move together, which affects diversification strategies and risk management.

Cov(X, Y) = Σ[(Xᵢ - ȳ)(Yᵢ - ẋ)] / (n - 1)
```

```python
'''Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium,large), and Material (wood, metal, plastic), 
perform label encoding using Python's scikit-learn library. Show your code and explain the output.'''

import pandas as pd
from sklearn.preprocessing import LabelEncoder
df = pd.DataFrame({'Color': ['red', 'green', 'blue'], 'Size':['small', 'medium','large'],'Material': ['wood', 'metal', 'plastic']})
encoder = LabelEncoder()
encoded_df = pd.DataFrame()
for column in df.columns:
    encoded_df[column] = encoder.fit_transform(df[column])

pd.concat([df,encoded_df],axis = 1)
```

```python
'''Q6.You are working on a machine learning project with a dataset containing several categorical variables, including "Gender" (Male/Female), "Education 
Level" (High School/Bachelor's/Master's/PhD), and "Employment Status" (Unemployed/Part-Time/Full-Time). Which encoding method would you use for each variable,
and why?'''





'''I will used One-hot-encoding for 'Gender' feature as it is categorize into two category which can be treated with binary value 1 and 0.
For "Education Level" ordinal encoding as is describe the ranking or order for each individual category and for "Employment Status" I will use ordinal encoding
as it is not represent any kind of order or ranking.'''


import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
df2 = pd.DataFrame({'Gender': ['Male','Female','Male','Female'], 'Education Level':['High School',"Bachelor's","Master's",'PhD'],
                    'Employment Status': ['Unemployed','Part-Time','Full-Time','Full-Time']})
one_hot_encoder = OneHotEncoder()
ordinal_encoder1 = OrdinalEncoder(categories = [['High School',"Bachelor's","Master's",'PhD']])
ordinal_encoder2 = OrdinalEncoder(categories = [['Unemployed','Part-Time','Full-Time','Full-Time']])
one_hot_encoded = one_hot_encoder.fit_transform(df2[['Gender']])
one_hot_encoded_df=pd.DataFrame(one_hot_encoded.toarray(),columns = one_hot_encoder.get_feature_names_out())
ordinal_encoded1 = ordinal_encoder1.fit_transform(df2[['Education Level']])
ordinal_encoded_df1 = pd.DataFrame(ordinal_encoded1, columns=['encoded_Education_Level'])
ordinal_encoded2 = ordinal_encoder2.fit_transform(df2[['Employment Status']])
ordinal_encoded_df2 = pd.DataFrame(ordinal_encoded2, columns=['encoded_Employment_status'])
pd.concat([one_hot_encoded_df,ordinal_encoded_df1,ordinal_encoded_df2],axis = 1)
```
