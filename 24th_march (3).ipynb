```python
"""Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?"""
#Import necessary libraries
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split


#lOAD DATASET
df3 = pd.read_csv('winequality-red.csv')


#See shape of data
print("shape : ",df3.shape)
""" 1592 rows and 12 columns are present in dataset"""


#VIEW SOME INITIAL DATAPOINTS
print(df3.head())

#NOW WE ARE GOING TO CHECK FOR MISSING VALUES
print(df3.isnull().sum())
""" There we can see no missing values are present in dataset ."""


#now see the over all information
print("Overall INformation of the dataset : ",df3.info())
"""From informaton we can observe that there are only float and integer values are present so need of encoding. """


#Now calculate the unique value of each feature.
for i in df3.columns.values:
    print("Unique value of  ",format(i),"is :" ,df3[i].nunique())
"""From the above code we can conclude that 'Density' have maximum number of unique values and 'quality' have minimum number of unique feature."""


#Now describe the all features
print(df3.describe())
"""We can conclude that from min to max percentile the difference is high of 'free sulfur dioxide' and  'total sulfur dioxide' feature."""


# check for duplicate data
print(df3[df3.duplicated()])
"""Here we can see duplicate datas  are present so remove all duplicate datas"""
df3.drop_duplicates(inplace = True)

#Checking Co-relation between features before PCA
plt.subplots(1,3,figsize=(25,8))
plt.subplot(121)
sns.heatmap(df3.corr(),annot = True)
"""1.density has a strong positive correlation with residual sugar, whereas it has a strong negative correlation with alcohol.
2.pH & fixed acidity has negative correlation.
3.density & fixed acidity has positive correlation.
4.citric acid & fixed acidity has positive correlation.
5.citric acid & volatile acidity has negative correlation.
6.free sulphur dioxide & total sulphur dioxide has positive correlation."""

#Standardize the features
scalar = StandardScaler()
scaled_data = pd.DataFrame(scalar.fit_transform(df3)) 
print(scaled_data)

#Applying Principal Component Analysis
pca = PCA(n_components = 12)
data_pca = pca.fit_transform(scaled_data)
data_pca = pd.DataFrame(data_pca,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12'])
data_pca.head()

#Checking Co-relation between features after PCA
plt.subplot(122)
sns.heatmap(data_pca.corr())
```
